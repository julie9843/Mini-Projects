## Task
What are the different modes of transport which are operated by Transfer for London? How many of modes do they have?

Print the list with different modes of transport, plus their count. Example output:
```
[bus, cable-car,.....]
Number of different modes of transport is: xyz
```

We need to search the documentation for correct request.

````python
url_mode="https://api.tfl.gov.uk/Journey/Meta/Modes"
mode_request=requests.get(url_mode)
````

````python
JSON(mode_request.json())
````

````python
mode_transport=[]
for x in mode_request.json():
    if x['isTflService']: 
        mode_transport.append(x['modeName'])
        
len(mode_transport)
print("Number of different modes of transport is: ", len(mode_transport))
````

````python
not_working=[]
for x in mode_request.json():
    if x['isTflService']==False : 
        not_working.append(x['modeName'])
len(not_working)
````

## Task
How many BikePoints in London are operated by Transfor for London? How many docks are in **all** BikePoints? There is the information for empty and full docks for each BikePoint.

````python
url_bikepoint="https://api.tfl.gov.uk/BikePoint/"
bikepoint_request = requests.get(url_bikepoint)
````

````python
JSON(bikepoint_request.json())
````

````python
len(bikepoint_request.json())

print("The number of bikepoints in London are ",len(bikepoint_request.json()))
````

````python
JSON(bikepoint_request.json())

functional_dock_list = [] 
broken_docks =[]
for bikepoint in bikepoint_request.json():
    num_bikes = int(bikepoint["additionalProperties"][6]['value'])
    num_empty = int(bikepoint["additionalProperties"][7]['value'])
    num_docks = int(bikepoint["additionalProperties"][8]['value'])
    broken_docks_l = num_docks - (num_bikes+num_empty)
    broken_docks.append(broken_docks_l)
    functional_docks = (num_docks - broken_docks_l)
    functional_dock_list.append(functional_docks)

print(sum(functional_dock_list))
print(sum(broken_docks))
````

## Task
Plan the journey from Heathrow Airport to Tower Bridge using Bus and Tube? Which way is faster? Example output:
```
Planned duration:
Bus: x minutes
Tube: y minutes
```

We need to search the documentation for correct requests and parameters we need.

````python
journey_url = "https://api.tfl.gov.uk/Journey/JourneyResults/Tw61AP/to/SE14TW&time&mode"
journey_request = requests.get(journey_url)
````

````python
# Time it takes by bus 
bus_url = "https://api.tfl.gov.uk/Journey/JourneyResults/Tw61AP/to/SE14TW&time&bus"
bus_request=requests.get(bus_url)
JSON(bus_request.json())
````

````python
bus_request.json()['journeys'][0]['duration']
````

````python
bus_time=[]
for c in bus_request.json()['journeys']:
    bus_time.append(c['duration'])
    
print(min(bus_time))
````

````python
# Time it takes by tube 
tube_url = "https://api.tfl.gov.uk/Journey/JourneyResults/Tw61AP/to/SE14TW&time&tube"
tube_request=requests.get(tube_url)
JSON(tube_request.json())
````

````python
tube_time=[]
for b in tube_request.json()['journeys']:
    tube_time.append(b['duration'])
print(min(tube_time))
````

````python
if min(tube_time)> min(bus_time):
    print("Tube takes longer than bus by", min(tube_time)-min(bus_time),"minutes")
elif min(tube_time)< min(bus_time):
    print("Bus takes longer than bus by", (min(bus_time)-min(tube_time),"minutes"))
else:
    print("Takes the same amount of time")

````
